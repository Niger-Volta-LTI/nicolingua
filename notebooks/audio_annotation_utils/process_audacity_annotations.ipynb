```python
from pathlib import Path
import csv
import pydub
from pydub.playback import play as play_audio
import IPython
```

```python
for fname in Path(BASE_DIR).rglob("*.txt"):
    # filter out OSX "._" files!
    if fname.stem.startswith("._"):
        continue
        
    with open(fname) as f:
        audio_fname = str(fname).replace(".txt", ".wav")
        if not Path(audio_fname).is_file():
            print(f"Skipping {fname}, for which no audio file could be found.")
            continue
            
        audio_segment = pydub.AudioSegment.from_file(audio_fname)
        reader = csv.DictReader(f, fieldnames = ["start", "end", "label"], delimiter="\t")
        for a in reader:
            start_ms = float(a["start"]) * 1000 - 100
            end_ms = float(a["end"]) * 1000 + 100
            segment = audio_segment[start_ms:end_ms]
            segment_fname = f"{fname.stem}_{a['label']}.wav" 
            segment.export(segment_fname)
            #play_audio(segment)
            # IPython.display.Audio(segment_fname)
            display(segment_fname, IPython.display.Audio(segment_fname))
            #break
```
